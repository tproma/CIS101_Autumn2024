```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0):
  print(f'{year} is leap year')
else:
   print(f'{year} is NOT a leap year')
```

```python
score = int(input("Enter a score:"))
if 90<= score <=100:
  grade = "A"
elif 80<= score <90:
  grade = "B"
elif 70<= score <80:
  grade = "C"
else:
  grade = "F"
print("Grade : " , grade)
```

```python
def Grade_checker(score):
  if 90<= score <=100:
    grade = "A"
  elif 80<= score <90:
    grade = "B"
  elif 70<= score <80:
    grade = "C"
  else:
    grade = "F"
  #print("Grade : " , grade)
  return grade
```

```python
student1 = Grade_checker(66)
print(student1)
```

```python
student3 = Grade_checker(100)
print(student3)
```

```python
def check_divisibility(num, check):
    if num % check == 0:
      print(num,'is divisible by',check)
    else:
      print(num,'is not divisible by',check)

num = int(input("Enter a number to check: "))
check = int(input("Enter a number to divide by: "))
check_divisibility(num, check)
```
