```python
year = int(input("Enter a year:"))

if (year % 4 == 0 and year % 100 != 0) :
  print(f'{year} is leap year')
else:
   print(f'{year} is NOT leap year')
```

```python
score = int(input("Enter a score:"))

if 90<=score<=100:
  grade = "A"
elif 80<=score<90:
  grade = "B"
elif 70<=score<80:
  grade = "C"
else:
  grade = "F"

print("Grade : " , grade)
```

```python
def Grade_ckecker(score):
  if 90<=score<=100:
    grade = "A"
  elif 80<=score<90:
    grade = "B"
  elif 70<=score<80:
    grade = "C"
  else:
    grade = "F"

 # print("Grade : " , grade)
  return grade
```
