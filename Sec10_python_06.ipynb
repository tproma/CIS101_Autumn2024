### For Loop

range(stop)

range(start, stop)

range(start, stop, step)

```python
for i in range(5):
  print(i)
```

```python
for i in range(5):
  print(i+2)
```

```python
for i in range(5):
  print(i+1, end = ",")
```

### range(start, stop)

```python
for i in range(11,21):
  print(i)
```

### range(start, stop, step)

```python
for i in range(1,10,2):
  print(i)
```

```python
for i in range(1,10,4):
  print(i)
```

### EVEN / ODD

```python
for i in range(0, 11, 2):
  print(i,end = "--")
```

```python
for i in range(1, 12, 2):
  print(i, end = "--")
```

### With Operators

```python
for i in range(0, 10):
  print(i, end = ",")
```

```python
for i in range(0, 10):
  if i%2 == 0:
    print(i, end = ",")
```

```python
for i in range(0, 10):
  if i%2 == 1:
    print(i, end = ",")
```

```python
for i in range(1, 20):
  if i%3 == 0:
    print(i, end = ",")
```

```python
sum = 0
for i in range(0, 11):
  print(i, end = "+")
  sum = sum +i
print("=",sum)
```

```python
fact = 1
for i in range(1,6):
  print(i, end = "*")
  fact = fact * i

print( "= ", fact)
```
