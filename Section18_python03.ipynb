```python
n = int(input())

if n> 100:
  print(f'{n} is greater than 100')
else:
  print(f'{n} is less than 100')
```

```python
# Character Check
char = input()
vowels = "aeiou"

if char in vowels:
  print(f'{char} is vowel')
else:
   print(f'{char} is consonant')
```
