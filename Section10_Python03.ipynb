```python
n = int(input())

if n<=10:
  print(f'{n} is less than or equal to 10')
else:
  print(f'{n}  greater than 10')
```

```python
# Character check

char = input()

vowels = "aeiou"

if char in vowels:
  print(f'{char} is vowel')
else:
  print(f'{char} is consonant')
```
